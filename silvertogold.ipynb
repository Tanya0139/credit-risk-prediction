```python
ip = "/mnt/silver/"
op = "/mnt/gold/"
```

```python
df = spark.read.format('parquet').option("header", "true").option("inferSchema", "true").load(ip)
display(df)
```

```python
import re

def clean_col_name(name):
    return re.sub(r'[ ,;{}()\n\t=]', '_', name)

df_clean = df.toDF(*[clean_col_name(c) for c in df.columns])
df_clean.write.format("delta").mode("overwrite").save("/mnt/gold/")
```
